## Importing necessary Library and load the data into pandas framework

```python
# importing necessary library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# reading data into the pandas dataframe
df=pd.read_csv('india-state-wise-data-analysis.csv')
```

```python
# Checking First five rows of the data 
df.head()
```

```python
# checking the shape of the data
df.shape
```

```python
# checking the information of data
'''like column name,
  type of column ,
  fill values
'''
df.info()
```

```python
#Checking the missing values or null entries

df.isnull().sum()
```

```python
#Decribing the statistic of the data

df.describe()
```

```python
# spliting the state and district form the State & District column

df[['District_code','state','District']]=df['State & District'].str.split(',', expand=True)

# Deleting state and district column
df.drop(['State & District'],axis=1,inplace=True)

# spliting the district code column into code and deleting district code column
df[['Dis','District_code']]=df['District_code'].str.split(':',expand=True)

# similarly spliting state column into the state

df[['State','state']]=df['state'].str.split(':',expand=True)

df.drop('State',axis=1,inplace=True)


# Spliting of district colmn into district

df[['Dis','District']]=df['District'].str.split(':',expand=True)

df.drop('Dis',axis=1,inplace=True)

```

## Sub Assignment 2: Exploring Indian States Data

```python
## "1. What is the population of each state?"

#Visvalization of the data

plt.figure(figsize=(10,8))
plt.xlabel("States")
plt.ylabel("Population")
plt.title('Bar plot of state and population')
plt.bar(df['state'],df['Population'])
plt.xticks(rotation=90)
```

```python
# "2. how are the 'Male' and 'Female' populations distributed across Indian states?

# Group the data by state_name and calculate the total Male and Female
grouped_data = df.groupby('state')[['Male', 'Female']].sum()

# Create a bar plot
plt.figure(figsize=(50, 10))
grouped_data.plot(kind='bar', stacked=True)
plt.title('Male and Female Distribution by Indian State')
plt.xlabel('State')
plt.ylabel('Male_and_Female')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group the data by state_name and calculate the total Male and Female
grouped_data = df.groupby('state')[['Male', 'Female']].sum()

# Create a bar plot
plt.figure(figsize=(14, 6))
grouped_data.plot(kind='bar', stacked=True)
plt.title('Male and Female Distribution by Indian State')
plt.xlabel('State')
plt.ylabel('Male_and_Female')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()

# Group the data by state_name and calculate the total Male and Female
grouped_data = df.groupby('state')[['Male', 'Female']].sum()

# Create a bar plot
plt.figure(figsize=(14, 6))
grouped_data.plot(kind='bar', stacked=True)
plt.title('Male and Female Distribution by Indian State')
plt.xlabel('State')
plt.ylabel('Male_and_Female')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# "3. how is the distribution of literate individuals across Indian states?"

plt.figure(figsize=(12, 6))
plt.title('Literate Distribution by Indian State')
plt.xlabel('State')
plt.ylabel('Literate')
plt.bar(df["state"],df["Literate"])
plt.xticks(rotation=90)
# Show the plot
plt.show()
```

```python
# Group the data by state and calculate the total 'Age_Group_0_29','Age_Group_30_49','Age_Group_50'
grouped_data = df.groupby('state')[['Age_Group_0_29','Age_Group_30_49','Age_Group_50']].sum()

# Create a bar graph
plt.figure(figsize=(16, 10))
grouped_data.plot(kind='bar', stacked=True)
plt.title('age distrubution by Indian State')
plt.xlabel('State')
plt.ylabel('age')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```
